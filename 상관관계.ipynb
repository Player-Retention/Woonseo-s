```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Destiny 2 리뷰 데이터 로드
file_path = "new reviews/Destiny_2_negative_reviews2.csv"  # 실제 파일 경로 입력
df = pd.read_csv(file_path, encoding="latin1")

# 'playtime' 컬럼을 숫자로 변환
df['playtime'] = pd.to_numeric(df['playtime'], errors='coerce')

# NaN 값 제거
df = df.dropna(subset=['playtime'])

# 플레이 시간 범주화
bins = [0, 10, 50, 100, 500, df['playtime'].max()]
labels = ['0-10h', '10-50h', '50-100h', '100-500h', '500h+']
df['play_time_category'] = pd.cut(df['playtime'], bins=bins, labels=labels, include_lowest=True)

# 각 플레이 시간 구간별 리뷰 개수 집계
playtime_counts = df['play_time_category'].value_counts().sort_index()

# 📊 막대 그래프 생성
plt.figure(figsize=(8, 5))
sns.barplot(x=playtime_counts.index, y=playtime_counts.values, palette="viridis")
plt.title("Destiny 2 - Number of Negative Reviews by Playtime")
plt.xlabel("Playtime Range")
plt.ylabel("Number of Reviews")
# plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Destiny 2 리뷰 데이터 로드
file_path = "new reviews/Warframe_positive_reviews2.csv"  # 실제 파일 경로 입력
df = pd.read_csv(file_path, encoding="latin1")

# 'playtime' 컬럼을 숫자로 변환
df['playtime'] = pd.to_numeric(df['playtime'], errors='coerce')

# NaN 값 제거
df = df.dropna(subset=['playtime'])

# 플레이 시간 범주화
bins = [0, 10, 50, 100, 500, df['playtime'].max()]
labels = ['0-10h', '10-50h', '50-100h', '100-500h', '500h+']
df['play_time_category'] = pd.cut(df['playtime'], bins=bins, labels=labels, include_lowest=True)

# 각 플레이 시간 구간별 리뷰 개수 집계
playtime_counts = df['play_time_category'].value_counts().sort_index()

# 📊 막대 그래프 생성
plt.figure(figsize=(8, 5))
ax = sns.barplot(x=playtime_counts.index, y=playtime_counts.values, palette="viridis")

# 각 막대 위에 값 표시
for i, value in enumerate(playtime_counts.values):
    ax.text(i, value + 2, str(value), ha='center', fontsize=9, color='black')

plt.title("Warframe - Number of Positive Reviews by Playtime")
plt.xlabel("Playtime Range")
plt.ylabel("Number of Reviews")
plt.show()
```
